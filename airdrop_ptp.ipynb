```python
import pandas as pd
import numpy as np

airdrop = pd.read_csv("./airdrop_ptp.csv", header = None)
airdrop.columns = ['rank', 'address']
airdrop.insert(2, 'amount', 0.0)
airdrop = airdrop.dropna()
airdrop['rank'] = airdrop['rank'].apply(lambda x:int(x))
```

```python
for ind in airdrop.index:
#      print(airdrop[0][ind], airdrop[1][ind])
    if airdrop['rank'][ind] == 1:
        airdrop['amount'][ind] = 15000
    elif airdrop['rank'][ind] == 2:
        airdrop['amount'][ind] = 8000
    elif airdrop['rank'][ind] == 3:
        airdrop['amount'][ind] = 3200
    elif airdrop['rank'][ind] <= 500:
        airdrop['amount'][ind] = 400
    elif airdrop['rank'][ind] <= 1000:
        airdrop['amount'][ind] = 150
```
